```python
import pandas as pd
import config
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
PATH_TO_DATA = config.data_csv_num
```

```python
data_movie = pd.read_csv(PATH_TO_DATA, index_col = 'Movie_Id')
```

```python
data_movie.head()
```

```python
data_movie.shape
```

```python
data_movie.dtypes
```

```python
sns.catplot(x='genre',y='rating',data=data_movie, kind='bar').set_xticklabels(rotation=30)
plt.show()
```

```python
sns.scatterplot(data=data_movie, x = "rating", y = "count", hue = 'genre')
plt.show()
```

```python
movie_genre_count = data_movie.groupby('genre', as_index = False).agg({'count':'count'}).sort_values('count', ascending = False)
```

```python
movie_genre_count.head()
```

```python
sns.catplot(x = 'genre', y = 'count', data = movie_genre_count, kind = 'bar').set_xticklabels(rotation=30)
plt.show()
```
